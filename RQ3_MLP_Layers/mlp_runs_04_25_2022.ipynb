```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
# This example supposes that the events file contains summaries with a
# summary value tag 'loss'.  These could have been added by calling
# `add_summary()`, passing the output of a scalar summary op created with
# with: `tf.scalar_summary(['loss'], loss_tensor)`.
bs_1024_loss = []
bs_1024_NDCG = []
bs_1024_HR = []
epochs = np.arange(200)
for e in tf.train.summary_iterator("events.out.tfevents.1650470460.42782dd5eb9b"):
    for v in e.summary.value:
      # print(v)
        if v.tag == 'model/loss': #or v.tag == 'performance/NDCG':
          bs_1024_loss.append(v.simple_value)
        elif v.tag == 'performance/NDCG':
          bs_1024_NDCG.append(v.simple_value)
        elif v.tag == 'performance/HR':
          bs_1024_HR.append(v.simple_value)
```
